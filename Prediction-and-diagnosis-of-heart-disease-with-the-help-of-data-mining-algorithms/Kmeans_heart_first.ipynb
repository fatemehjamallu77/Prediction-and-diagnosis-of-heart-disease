```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import scipy.stats as stats
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:/Users/lenovo/Desktop/heart.csv' )
dataset=data.to_numpy()
class_target=dataset[:,-1]
class_instances=dataset[:,:-1]
class_instances=stats.zscore(class_instances, axis=0)
```

```python
kmeans_heart = KMeans(n_clusters = 2)
kmeans_heart = kmeans_heart.fit(class_instances)
cluster=kmeans_heart.labels_
```

```python
print('centroids:\n', kmeans_heart.cluster_centers_)
print('clusters:\n', kmeans_heart.labels_)
```

```python
acc = accuracy_score(class_target, cluster)
print("accuracy:",acc)
print('silhouette:\n', silhouette_score(class_instances, kmeans_heart.labels_))
```

```python
cm = confusion_matrix(class_target, cluster)
print('confusion_matrix:')
print(cm)
```

```python
print('Precision: %.3f' % precision_score(class_target, cluster))
```

```python
print('Recall: %.3f' % recall_score(class_target, cluster))
```

```python
print('F1 Score: %.3f' % f1_score(class_target, cluster))
```

```python
two_d_data = class_instances[:, :2]
kmeans_heart = kmeans_heart.fit(two_d_data)

plt.scatter(two_d_data[:, 0], two_d_data[:, 1], c = kmeans_heart.labels_, s=250)
plt.scatter(kmeans_heart.cluster_centers_[:, 0], kmeans_heart.cluster_centers_[:, 1], c = 'k', marker = 'X', s=150)
```

```python
fig = plt.figure() 
ax = Axes3D(fig)
KMeans_heart = KMeans(n_clusters = 2)
three_d_data = class_instances[:, :3]
KMeans_heart = KMeans_heart.fit(three_d_data)

plt.scatter(three_d_data[:, 0], three_d_data[:, 1], three_d_data[:, 2], c = KMeans_heart.labels_)
```

```python
print('silhouette:\n', silhouette_score(three_d_data, KMeans_heart.labels_))
```
